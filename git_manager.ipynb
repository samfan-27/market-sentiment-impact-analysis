```python
# === CELL 4: PUSH (End of Work) ===
# 1. Check what changed
print("--- Status ---")
!git status

# 2. Ask if you want to proceed
choice = input("\nDo you want to push these changes? (y/n): ")

if choice.lower() == 'y':
    # 3. Stage everything
    !git add .

    # 4. Commit with a custom message
    msg = input("Enter commit message: ")
    !git commit -m "{msg}"

    # 5. Push safely (No force!)
    !git push origin main
    print("\nSent to GitHub!")
else:
    print("\nPush cancelled.")
```
